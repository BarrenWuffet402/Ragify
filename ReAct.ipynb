```python
from datetime import datetime, timedelta

import dspy
```

```python
dspy.configure(lm=dspy.LM("openai/gpt-5.4-nano"))
```

```python
def get_current_datetime() -> str:
    """Get the current date and time as a string."""
    return datetime.now().isoformat()
```

```python
def datetime_calculator(
    start_time: str,
    delta_minutes: int,
) -> str:
    """Calculate a new datetime by adding minutes (including negative values) to a given start time."""
    start_dt = datetime.fromisoformat(start_time)
    new_dt = start_dt + timedelta(minutes=delta_minutes)
    return new_dt.isoformat()
```

```python
sig = dspy.Signature(
    "question -> answer: str",
    instructions="Answer the question using dates and times.",
)
```

```python
react_agent = dspy.ReAct(sig, tools=[get_current_datetime, datetime_calculator])
# react_agent
```

```python
result = react_agent(question="I was born in June 1989. How many years do I have until I can retire if I want to retire at 67?")
result
```
